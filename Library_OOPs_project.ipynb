```python
class Library:# class attribute
    # Instance attribute
    def __init__(self,list,name):
        self.booklist=list #1
        self.name=name  #2
        self.lendDict={}  #3
        # in 1,2,3 we created 3 instance attribute
        
    def displaybooks(self):
        print(f"we have following books in library:{self.name}")
        for book in self.booklist:
            print(book)
    def lendbook(self,user,book):
        if book not in self.lendDict.keys():
            self.lendDict.update({book:user})
            print("lender book info updated, now you can take book")
        else:
            print(f"book already in use by  {self.lendDict[book]}")
    def addbook(self,book):
        self.booklist.append(book)
        print("book added to booklist")
    def returnbook(self,book):
        self.lendDict.pop(book)
            
# object
if __name__=="__main__":
    ram=Library(["ML basics","biosphere"],"Karan")
    while (True):
        print(f"welcome to {ram.name} library")
        print("1. for dispalybooks")
        print("2. for lend a book ")
        print("3 add a book")
        print("4 return a book")

        user_input=input("enter no in range 1,2,3,4\n")
        if user_input not in ['1','2',"3","4"]:
            print("plz enter valid input")
            continue
        else:
            user_input=int(user_input)
            
        if user_input==1:
            ram.displaybooks()
        elif user_input==2:
            book=input("enter book name you want to lend")
            user=input("enter your name")
            ram.lendbook(user,book)
        elif user_input==3:
            book=input("enter book you want to add")
            ram.addbook(book)
        elif user_input==4:
            book=input("enter book you want to retrun")
            ram.returnbook(book)
            
            
        else :
            print("not a valid input")
            
        print("q to quit and c to continue\n")
        user_input2=""
        while(user_input2!="c" and user_input2!="q"):
            
            user_input2=input("")
            
            if user_input2=="c":
                continue
            elif user_input2=="q":
               
                print("thank you")
    exit()
            
            
#         exit()
#             else:
#                 print(" invalid options please enter only  q/c")

        

```
